```python
import tensorflow as tf

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Flatten, Dense
from tensorflow.keras.optimizers import SGD, Adam

import numpy as np
```

## 데이터셋 생성

```python
# [1] 데이터셋 생성

loaded_data = np.loadtxt('./sps.csv', delimiter=',', dtype=np.float32)

x_data = loaded_data[ :, 1:]
t_data = loaded_data[ :, [0]]

# 데이터 차원 및 shape 확인
print("loaded_data.ndim = ", loaded_data.ndim, ", loaded_data.shape = ", loaded_data.shape)
print("x_data.ndim = ", x_data.ndim, ", x_data.shape = ", x_data.shape)
print("t_data.ndim = ", t_data.ndim, ", t_data.shape = ", t_data.shape)
```

### W,b 초기화 단계

```python
# np.random.seed(0)

# W = np.random.rand(4,1)
# b = np.random.rand(1)
```

```python
## tensorflow 이용

model = Sequential()
model.add(Dense(1,input_shape=(4,), activation='linear')) # 입력층 + 출력층
```

```python
print(model.weights )
```

## loss함수 정의 미분방법 정의 // 모델 컴파일

```python
# def loss_func(x,t):
#   y = np.dot(x,W) + b
#   return(np.sum((t-y)**2))/(len(x)) # MSE
```

```python
# 모델 컴파일 및 summary
model.compile(optimizer=SGD(learning_rate=1e-2), loss='mse')

model.summary()
```

## 모델 학습 //model.fit

```python
hist= model.fit(x_data,t_data, epochs=2000)
```

```python
print(model.weights) # 학습후 W,b상황
```
